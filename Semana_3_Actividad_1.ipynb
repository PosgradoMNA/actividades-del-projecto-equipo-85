```python
import pandas as pd
import numpy as np
```

```python
#Importar base
filename = "https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv"
df = pd.read_csv(filename)
df.head()
```

```python
#Cambiar el nombre a las columnas para identificar a que corresponden los datos
df.columns = ['ID','Motivocredito','Genero','Educacion','EstadoCivil','Edad',
              'Pagos_Sep_2005','Pagos_Ago_2005','Pagos_Jul_2005','Pagos_Jun_2005','Pagos_May_2005','Pagos_Abr_2005',
              'Estado_Sep_2005','Estado_Ago_2005','Estado_Jul_2005','Estado_Jun_2005','Estado_May_2005','Estado_Abr_2005',
              'Monto_Sep_2005','Monto_Ago_2005','Monto_Jul_2005','Monto_Jun_2005','Monto_May_2005','Monto_Abr_2005','Y']
df.head()
```

```python
##df.isnull().values.any() #Identificar si la base tiene espacios vacios
##df.isnull().any()        #Identificar por columna los espacios vacios
df.isnull().sum()        #Contar cuantos datos por columna están vacios
```

#Revisar si los valores de cada columna hacen sentido

```python
#Género (1 = masculino; 2 = femenino). Está ok
print(df['Genero'].unique())
print(df['Genero'].value_counts())
#Sustituiremos el único registo que esta en blanco por "2", ya que representa el 60% de los datos
df['Genero'].fillna(2, inplace = True) 
print(df['Genero'].unique())
```

```python
#Educación (1 = posgrado; 2 = universidad; 3 = secundaria; 4 = otros). Los que se encuentran en 5,6,0,nan los convertiremos en 4, para reflejarlo en "otros"
print(df['Educacion'].unique())
df.Educacion.replace({5: 4, 6: 4, 0: 4}, inplace=True)
df['Educacion'].fillna(4, inplace = True) 
print(df['Educacion'].unique())
```

```python
#Estado civil (1 = casado; 2 = soltero; 3 = otros). Los NA los cambiaremos por otros
print(df['EstadoCivil'].unique())
df.EstadoCivil.replace({0: 3}, inplace=True)
df['EstadoCivil'].fillna(3, inplace = True)
print(df['EstadoCivil'].unique())
```

```python
#Edad (año)
print(df['Edad'].unique())
#Utilizaremos el promedio de edad de las personas para susituir los NA
df.Edad.fillna(df.Edad.mean(), inplace=True)
print(df['Edad'].unique())
```

```python
#Para el caso de los pagos y estados de cuenta, los dejamos de considerar para el análisis
df.dropna(inplace = True)  
df.isnull().sum()
```
